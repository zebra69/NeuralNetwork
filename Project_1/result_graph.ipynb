```python
import matplotlib.pyplot as plt
import numpy as np
```

## Part1

```python
w_delta = [6.281398, 6.277880, 9.502765]

w_perceptron = [2.432057, 1.198480, 2.729427 ]
```

```python
x = np.arange(-2, 2, 0.1)

y_delta = (-w_delta[0]*x + w_delta[2])/w_delta[1]
y_perceptron = (-w_perceptron[0]*x + w_perceptron[2])/w_perceptron[1]

ax1 = plt.subplot(1,2,1)
ax1.set_title('delta')
ax1.plot(x, y_delta)
ax1.plot(0,0,marker='.',color='blue')
ax1.plot(0,1,marker='.',color='blue')
ax1.plot(1,0,marker='.',color='blue')
ax1.plot(1,1,marker='.',color='red')
ax2 = plt.subplot(1,2,2)
ax2.set_title('perceptron')
ax2.plot(x, y_perceptron)
ax2.plot(0,0,marker='.',color='blue')
ax2.plot(0,1,marker='.',color='blue')
ax2.plot(1,0,marker='.',color='blue')
ax2.plot(1,1,marker='.',color='red')
plt.show()
```

## Part2

```python
w_delta = [[2.637271, 2.113399, 1.789992],
          [-0.219082, -1.095347, -0.031215],
          [-0.940954, 1.158520, 0.511674]]

w_perceptron = [[7.794171, 6.921058, -0.133537],
               [-2.704680, -11.502300, 1.239326],
               [-0.327217, 0.299520, 0.290121]]
```

```python
x = np.arange(-15, 15, 0.1)
y_delta = []
y_perceptron = []
for i in range(0,3):
    y_delta.append((-w_delta[i][0]*x + w_delta[i][2])/w_delta[i][1])
    y_perceptron.append((-w_perceptron[i][0]*x + w_perceptron[i][2])/w_perceptron[i][1])

ax1 = plt.subplot(1,2,1)
ax1.set_title('delta')
ax1.plot(x, y_delta[0])
ax1.plot(x, y_delta[1])
ax1.plot(x, y_delta[2])
ax1.plot(10,2,marker='.',color='blue')
ax1.plot(2,-5,marker='.',color='blue')
ax1.plot(-5,5,marker='.',color='blue')

ax2 = plt.subplot(1,2,2)
ax2.set_title('perceptron')
ax2.plot(x, y_perceptron[0])
ax2.plot(x, y_perceptron[1])
ax2.plot(x, y_perceptron[2])
ax2.plot(10,2,marker='.',color='blue')
ax2.plot(2,-5,marker='.',color='blue')
ax2.plot(-5,5,marker='.',color='blue')

plt.show()
```
